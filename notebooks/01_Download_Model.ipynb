```python
import httpx
import platformdirs
import zipfile
from pathlib import Path
from tqdm.notebook import tqdm

cache_dir = Path(platformdirs.user_cache_dir("katago-onnx"))
cache_dir.mkdir(parents=True, exist_ok=True)

network_name = "kata1-b28c512nbt-adam-s11165M-d5387M"
torch_checkpoint_url = f"https://media.katagotraining.org/uploaded/networks/zips/kata1/{network_name}.zip"
```

```python
zip_path = cache_dir / f"{network_name}.zip"
model_dir = cache_dir / network_name

print(f"Downloading {network_name}...")
with httpx.stream("GET", torch_checkpoint_url, follow_redirects=True) as response:
    response.raise_for_status()
    total = int(response.headers.get("Content-Length", 0))

    with open(zip_path, "wb") as f, tqdm(total=total, unit="B", unit_scale=True, unit_divisor=1024) as pbar:
        for chunk in response.iter_bytes():
            f.write(chunk)
            pbar.update(len(chunk))

print("Unzipping...")
model_dir.mkdir(exist_ok=True)
with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(model_dir)
```
